```python

# Practical Examples:

# 1. Name Validation
def validate_name(name):
    """Check if name contains only letters and spaces"""
    return all(char.isalpha() or char.isspace() for char in name)

# 2. Username Validation
def validate_username(username):
    """Check if username contains only letters and numbers"""
    return username.isalnum()

# 3. Phone Number Validation
def validate_phone(phone):
    """Check if phone number contains only digits"""
    return phone.isdigit() and len(phone) == 10

# 4. Title Case Checker
def is_proper_title(title):
    """Check if string is properly title-cased"""
    return title.istitle()

# Testing the validation functions
print("\n--- Practical Examples ---")
# Name validation
names = ["John Doe", "John123", "Mary Smith", "Bob!"]
for name in names:
    print(f"Name '{name}' is {'valid' if validate_name(name) else 'invalid'}")

# Username validation
usernames = ["john123", "mary_smith", "peter42!", "admin123"]
for username in usernames:
    print(f"Username '{username}' is {'valid' if validate_username(username) else 'invalid'}")

# Phone number validation
phones = ["1234567890", "123-456-7890", "12345", "abcdefghij"]
for phone in phones:
    print(f"Phone '{phone}' is {'valid' if validate_phone(phone) else 'invalid'}")

# Title validation
titles = ["The Lord of the Rings", "Star wars", "Harry Potter", "THE MATRIX"]
for title in titles:
    print(f"Title '{title}' is {'properly' if is_proper_title(title) else 'improperly'} formatted")

# Combined Validation Example
def validate_form(name, username, phone):
    """Validate a complete form"""
    errors = []
    if not validate_name(name):
        errors.append("Invalid name")
    if not validate_username(username):
        errors.append("Invalid username")
    if not validate_phone(phone):
        errors.append("Invalid phone number")
    return errors
```

```python
# Demonstrating mutable vs immutable objects
import sys

# Immutable Objects Example (Strings)
text = "hello"
print(f"Original text: {text}, ID: {id(text)}")
text += " world"  # Creates a new string object
print(f"Modified text: {text}, ID: {id(text)}")  # Different ID!

# Mutable Objects Example (Lists)
numbers = [1, 2, 3]
print(f"\nOriginal list: {numbers}, ID: {id(numbers)}")
numbers.append(4)  # Modifies the same list object
print(f"Modified list: {numbers}, ID: {id(numbers)}")  # Same ID!

# Memory size of objects
print(f"\nMemory size of string '{text}': {sys.getsizeof(text)} bytes")
print(f"Memory size of list {numbers}: {sys.getsizeof(numbers)} bytes")

# Reference sharing with mutable objects
list1 = [1, 2, 3]
list2 = list1  # Both variables reference the same list object
list1.append(4)
print(f"\nlist1: {list1}")
print(f"list2: {list2}")  # list2 also shows the change
print(f"Same object? {list1 is list2}")

# Creating a copy instead of sharing reference
list3 = list1.copy()
list1.append(5)
print(f"\nlist1: {list1}")
print(f"list3: {list3}")  # list3 remains unchanged
print(f"Same object? {list1 is list3}")

# Immutable objects and references
x = 42
y = x  # For immutable objects, this creates a new reference
x = 43  # Creates a new object, doesn't affect y
print(f"\nx: {x}")
print(f"y: {y}")  # y remains unchanged
```

```python
# Inheritance Example
class Vehicle:
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year
    
    def start(self):
        return f"The {self.year} {self.make} {self.model} is starting."

class Car(Vehicle):
    def __init__(self, make, model, year, num_doors):
        super().__init__(make, model, year)  # Call parent class constructor
        self.num_doors = num_doors
    
    def honk(self):
        return "Beep beep!"

class Motorcycle(Vehicle):
    def __init__(self, make, model, year, has_sidecar):
        super().__init__(make, model, year)
        self.has_sidecar = has_sidecar
    
    def wheelie(self):
        return "Performing a wheelie!"

# Creating vehicle objects
car = Car("Toyota", "Camry", 2022, 4)
motorcycle = Motorcycle("Harley-Davidson", "Sportster", 2021, False)

print("\nVehicle examples:")
print(car.start())
print(car.honk())
print(motorcycle.start())
print(motorcycle.wheelie())
```

```python

# Encapsulation Example
class BankAccount:
    def __init__(self, account_holder, initial_balance):
        self.__holder = account_holder  # Private attribute
        self.__balance = initial_balance  # Private attribute
    
    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            return f"Deposited ${amount}. New balance: ${self.__balance}"
        return "Invalid deposit amount"
    
    def get_balance(self):
        return f"Current balance: ${self.__balance}"

# Using the BankAccount class
print("\nBank account example:")
account = BankAccount("Oliver", 1000)
print(account.get_balance())
print(account.deposit(500))
# print(account.__balance)  # This would raise an AttributeError
```

```python
# Polymorphism Example
class Animal:
    def speak(self):
        pass

class Dog(Animal):
    def speak(self):
        return "Woof!"

class Cat(Animal):
    def speak(self):
        return "Meow!"

class Duck(Animal):
    def speak(self):
        return "Quack!"

# Function demonstrating polymorphism
def animal_sound(animal):
    return animal.speak()

# Creating animal objects
dog = Dog()
cat = Cat()
duck = Duck()

print("\nPolymorphism example:")
print(f"Dog says: {animal_sound(dog)}")
print(f"Cat says: {animal_sound(cat)}")
print(f"Duck says: {animal_sound(duck)}")
```

```python
from abc import ABC, abstractmethod
import math

# Abstract Shape class
class Shape(ABC):
    @abstractmethod
    def area(self):
        pass
    
    @abstractmethod
    def perimeter(self):
        pass

# Concrete implementation of Shape
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius
    
    def area(self):
        return math.pi * self.radius ** 2
    
    def perimeter(self):
        return 2 * math.pi * self.radius

class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height
    
    def area(self):
        return self.width * self.height
    
    def perimeter(self):
        return 2 * (self.width + self.height)

# Using the shapes
circle = Circle(5)
rectangle = Rectangle(4, 6)

print(f"Circle area: {circle.area():.2f}")
print(f"Circle perimeter: {circle.perimeter():.2f}")
print(f"Rectangle area: {rectangle.area()}")
print(f"Rectangle perimeter: {rectangle.perimeter()}")

# This would raise an error:
# shape = Shape()  # Can't instantiate abstract class

# Another example with a more complex interface
class PaymentProcessor(ABC):
    @abstractmethod
    def process_payment(self, amount):
        pass
    
    @abstractmethod
    def refund_payment(self, amount):
        pass

class CreditCardProcessor(PaymentProcessor):
    def process_payment(self, amount):
        return f"Processing ${amount} via credit card"
    
    def refund_payment(self, amount):
        return f"Refunding ${amount} to credit card"

class PayPalProcessor(PaymentProcessor):
    def process_payment(self, amount):
        return f"Processing ${amount} via PayPal"
    
    def refund_payment(self, amount):
        return f"Refunding ${amount} to PayPal account"

# Using payment processors
cc_processor = CreditCardProcessor()
paypal_processor = PayPalProcessor()

print("\nPayment processing examples:")
print(cc_processor.process_payment(100))
print(cc_processor.refund_payment(50))
print(paypal_processor.process_payment(75))
print(paypal_processor.refund_payment(25))
```

```python
import time

def compare_search_performance(size=10000, search_value=None):
    try:
        # Create large lists and sets for comparison
        large_list = list(range(size))
        large_set = set(range(size))
        
        # If no search value specified, use the last element
        if search_value is None:
            search_value = size - 1

        # Validate search value
        if search_value >= size:
            raise ValueError(f"Search value {search_value} must be less than size {size}")

        # Time list search
        start = time.time()
        found_in_list = search_value in large_list
        list_time = time.time() - start

        # Time set search
        start = time.time()
        found_in_set = search_value in large_set
        set_time = time.time() - start

        # Print results
        print("\nPerformance Comparison:")
        print(f"Data structure size: {size}")
        print(f"Searching for value: {search_value}")
        print(f"Found in list: {found_in_list}")
        print(f"Found in set: {found_in_set}")
        print(f"Time to find in list: {list_time:.6f} seconds")
        print(f"Time to find in set: {set_time:.6f} seconds")
        
        # Avoid division by zero
        if set_time > 0:
            speed_difference = list_time/set_time
            print(f"Set is approximately {speed_difference:.0f}x faster")
        else:
            print("Set search was too fast to measure")

    except Exception as e:
        print(f"An error occurred: {str(e)}")

# Test with different scenarios
print("Test 1: Original case")
compare_search_performance(10000, 9999)

print("\nTest 2: Smaller dataset")
compare_search_performance(1000, 999)

print("\nTest 3: Search for first element")
compare_search_performance(10000, 0)

print("\nTest 4: Search for non-existent value")
compare_search_performance(10000, -1)
```

```python
# Conditional Statements
print("Conditional Statements Examples:")

# Basic if-elif-else
def grade_score(score):
    if score >= 90:
        return 'A'
    elif score >= 80:
        return 'B'
    elif score >= 70:
        return 'C'
    elif score >= 60:
        return 'D'
    else:
        return 'F'

print(f"Score 85 gets grade: {grade_score(85)}")
print(f"Score 92 gets grade: {grade_score(92)}")
print(f"Score 65 gets grade: {grade_score(65)}")

# Ternary operator
age = 20
status = "adult" if age >= 18 else "minor"
print(f"\nStatus: {status}")

print("\nLoop Examples:")
# For Loops
# Basic for loop
print("Counting:")
for i in range(5):
    print(i, end=' ')
print()  # New line

# Loop over a list
fruits = ["apple", "banana", "orange"]
print("\nFruits:")
for fruit in fruits:
    print(fruit, end=' ')
print()

# Loop with enumerate
print("\nEnumerated fruits:")
for index, fruit in enumerate(fruits):
    print(f"{index}: {fruit}")

# While Loops
print("\nWhile loop example:")
count = 0
while count < 5:
    print(count, end=' ')
    count += 1
print()

# Break and Continue
print("\nBreak example:")
for i in range(10):
    if i == 5:
        break
    print(i, end=' ')
print()

print("\nContinue example:")
for i in range(5):
    if i == 2:
        continue
    print(i, end=' ')
print()

# Loop with else
print("\nLoop with else:")
for i in range(3):
    print(i, end=' ')
else:
    print("\nLoop completed normally")

# Break with else
print("\nLoop with break and else:")
for i in range(3):
    if i == 1:
        break
    print(i, end=' ')
else:
    print("\nThis won't print because of break")
print()

print("\nException Handling Examples:")
# Exception Handling
def divide(a, b):
    try:
        result = a / b
    except ZeroDivisionError:
        return "Cannot divide by zero"
    except TypeError:
        return "Invalid types for division"
    else:
        return f"Result: {result}"
    finally:
        print("Division attempt completed")

print(divide(10, 2))    # Normal division
print(divide(10, 0))    # Division by zero
print(divide(10, "2"))  # Type error

# Custom Exception
class NegativeNumberError(Exception):
    """Raised when a negative number is encountered"""
    pass

def process_positive_number(number):
    try:
        if number < 0:
            raise NegativeNumberError("Number cannot be negative")
        return f"Processing {number}"
    except NegativeNumberError as e:
        return str(e)

print("\nCustom exception examples:")
print(process_positive_number(5))
print(process_positive_number(-5))
```

```python
import os
import shutil
from pathlib import Path

# File Operations
print("File Operations Examples:")

# Writing to a file
with open('example.txt', 'w') as f:
    f.write('Hello, World!\n')
    f.write('This is a test file.\n')
    f.write('Python is awesome!')

# Reading from a file
print("\nReading entire file:")
with open('example.txt', 'r') as f:
    content = f.read()
    print(content)

# Reading line by line
print("\nReading line by line:")
with open('example.txt', 'r') as f:
    for line in f:
        print(line.strip())

# Appending to a file
with open('example.txt', 'a') as f:
    f.write('\nThis line was appended!')

# Reading with different modes
print("\nReading with different modes:")
# Read all lines into a list
with open('example.txt', 'r') as f:
    lines = f.readlines()
    print("Number of lines:", len(lines))
    print("First line:", lines[0].strip())

print("\nOS Operations Examples:")
# OS Operations
# Current working directory
print("Current directory:", os.getcwd())

# Create a new directory
os.makedirs('test_dir', exist_ok=True)
print("Created directory: test_dir")

# List directory contents
print("\nDirectory contents:", os.listdir('.'))

# Path manipulation
file_path = os.path.join('test_dir', 'test.txt')
print("Joined path:", file_path)

# Create a file in the new directory
with open(file_path, 'w') as f:
    f.write('Test content')

# File information
print("\nFile information:")
print("File exists?", os.path.exists(file_path))
print("Is it a file?", os.path.isfile(file_path))
print("Is it a directory?", os.path.isdir('test_dir'))

# File size
print("File size:", os.path.getsize(file_path), "bytes")

# File timestamps
print("Last modified:", os.path.getmtime(file_path))

# Using pathlib (modern approach)
print("\nPathlib examples:")
path = Path('test_dir')
print("Path exists?", path.exists())
print("Is directory?", path.is_dir())
print("Parent directory:", path.parent)

# List all text files in directory
print("\nText files in directory:")
for file in path.glob('*.txt'):
    print(file)

# Clean up
print("\nCleaning up...")
os.remove(file_path)  # Remove test file
os.rmdir('test_dir')  # Remove test directory
os.remove('example.txt')  # Remove example file
print("Cleanup completed")
```
